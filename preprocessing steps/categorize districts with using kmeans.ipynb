```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
aydin_df = pd.read_excel('aydin.xlsx') 
mugla_df = pd.read_excel('mugla.xlsx')
denizli_df = pd.read_excel('denizli.xlsx')
```

```python
category_percantage = aydin_df.groupby(['İLÇE (3B)', 'Kesinti_Nedeni_Kategori']).size().unstack(fill_value=0)
category_percantage = category_percantage.div(category_percantage.sum(axis=1), axis=0)
```

```python
category_percantage
```

```python
def categorize(df):
    
    category_percantage = df.groupby(['İLÇE (3B)', 'Kesinti_Nedeni_Kategori']).size().unstack(fill_value=0)
    category_percantage = category_percantage.div(category_percantage.sum(axis=1), axis=0)

    print(category_percantage)

    kmeans = KMeans(n_clusters=4, random_state=42)
    category_of_district = kmeans.fit_predict(category_percantage)

    category_percantage['İlçe_Kategori'] = category_of_district
    df = df.merge(category_percantage['İlçe_Kategori'], left_on='İLÇE (3B)', right_index=True)

    #print(category_percantage)
    print(df[['İLÇE (3B)', 'İlçe_Kategori']])
    
    return df
    
```

```python
aydin_df = categorize(aydin_df)
```

```python
aydin_df
```

```python
mugla_df = categorize(mugla_df)
denizli_df = categorize(denizli_df)
```

```python
def new_index(df):
    column_to_move = 'İlçe_Kategori'
    new_index = 4
    column = df.pop(column_to_move)
    df.insert(new_index, column_to_move, column)

    return df
```

```python
aydin_df = new_index(aydin_df)
mugla_df = new_index(mugla_df)
denizli_df = new_index(denizli_df)
```
