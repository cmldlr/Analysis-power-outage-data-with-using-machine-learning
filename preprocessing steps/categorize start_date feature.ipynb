```python
import pandas as pd
```

```python
aydin_df = pd.read_excel('aydin_v1.xlsx')
denizli_df = pd.read_excel('denizli_v1.xlsx')
mugla_df = pd.read_excel('mugla_v1.xlsx')
```

```python
aydin_df['KESİNTİ BAŞLAMA TARİHİ VE ZAMANI (6)']
```

```python
#09-12 category : 1
#12-14 category : 2
#14-18 category : 3
#18-00 category : 4
#00-09 category : 5

def categorize_hour(hour):
    if 9 <= hour < 12:
        return 1
    elif 12 <= hour < 14:
        return 2
    elif 14 <= hour < 18:
        return 3
    elif 18 <= hour < 24:
        return 4
    else:
        return 5

```

```python
aydin_df['KESİNTİ BAŞLAMA TARİHİ VE ZAMANI (6)'] = pd.to_datetime(aydin_df['KESİNTİ BAŞLAMA TARİHİ VE ZAMANI (6)'])
aydin_df['Saat_Kategori'] = aydin_df['KESİNTİ BAŞLAMA TARİHİ VE ZAMANI (6)'].dt.hour.apply(categorize_hour)
```

```python
mugla_df['KESİNTİ BAŞLAMA TARİHİ VE ZAMANI (6)'] = pd.to_datetime(mugla_df['KESİNTİ BAŞLAMA TARİHİ VE ZAMANI (6)'])
mugla_df['Saat_Kategori'] = mugla_df['KESİNTİ BAŞLAMA TARİHİ VE ZAMANI (6)'].dt.hour.apply(categorize_hour)
denizli_df['KESİNTİ BAŞLAMA TARİHİ VE ZAMANI (6)'] = pd.to_datetime(denizli_df['KESİNTİ BAŞLAMA TARİHİ VE ZAMANI (6)'])
denizli_df['Saat_Kategori'] = denizli_df['KESİNTİ BAŞLAMA TARİHİ VE ZAMANI (6)'].dt.hour.apply(categorize_hour)
```

```python
desired_columns = ['KESİNTİ BAŞLAMA TARİHİ VE ZAMANI (6)', 'Saat_Kategori']
print(aydin_df[desired_columns])
```

```python
def new_index(df):
    column_to_move = 'Saat_Kategori'
    new_index = 14
    column = df.pop(column_to_move)
    df.insert(new_index, column_to_move, column)

    return df
```

```python
aydin_df = new_index(aydin_df)
mugla_df = new_index(mugla_df)
denizli_df = new_index(denizli_df)
```
