```python
import pandas as pd
import numpy as py 
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
aydin_df = pd.read_excel('aydin_v2.xlsx')
denizli_df = pd.read_excel('denizli_v2.xlsx')
mugla_df = pd.read_excel('mugla_v2.xlsx')
```

```python
column_data = aydin_df['ŞEBEKE UNSURU TİPİ (3C)']
value_counts = column_data.value_counts()
print(value_counts)
count_unique_values = column_data.nunique()
print(count_unique_values)
```

```python
#frequency encoding for 'ŞEBEKE UNSURU TİPİ (3C)
def frequency_encoding(df):
    
    freq_encoding = df['ŞEBEKE UNSURU TİPİ (3C)'].value_counts(normalize = True)
    df['ŞEBEKE UNSURU TİPİ freq_encoded'] = df['ŞEBEKE UNSURU TİPİ (3C)'].map(freq_encoding)
    
    column_to_move = 'ŞEBEKE UNSURU TİPİ freq_encoded'
    new_index = 6
    column = df.pop(column_to_move)
    df.insert(new_index, column_to_move, column)

    return df

aydin_df = frequency_encoding(aydin_df)
mugla_df = frequency_encoding(mugla_df)
denizli_df = frequency_encoding(denizli_df)
```

```python
column_data = aydin_df['SÜREYE GÖRE (5B)']
value_counts = column_data.value_counts()
print(value_counts)
count_unique_values = column_data.nunique()
print(count_unique_values)
```

```python
column_data = aydin_df['SEBEBE GÖRE (5C)']
value_counts = column_data.value_counts()
print(value_counts)
count_unique_values = column_data.nunique()
print(count_unique_values)
```

```python
column_data = aydin_df['BİLDİRİME GÖRE (5D)']
value_counts = column_data.value_counts()
print(value_counts)
count_unique_values = column_data.nunique()
print(count_unique_values)
```

```python
def one_hot_encoding(df):
    
   columns_to_encode = ['KAYNAĞA GÖRE (5A)', 'SÜREYE GÖRE (5B)', 'SEBEBE GÖRE (5C)', 'BİLDİRİME GÖRE (5D)']

   one_hot_encoder = OneHotEncoder()
   encoded_features = one_hot_encoder.fit_transform(df[columns_to_encode]).toarray()

   encoded_feature_names = one_hot_encoder.get_feature_names_out(columns_to_encode)

   encoded_df = pd.DataFrame(encoded_features, columns=encoded_feature_names)
   df = pd.concat([df, encoded_df], axis=1).drop(columns=columns_to_encode)

   return df
    
aydin_df = one_hot_encoding(aydin_df)
denizli_df = one_hot_encoding(denizli_df)
mugla_df = one_hot_encoding(mugla_df)
```
