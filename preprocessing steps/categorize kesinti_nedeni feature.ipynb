```python
def classify_cause(cause):
    #teknik arızalar
    technical_failures = [
       'OG Fider Açması', 'AG Sigorta Atığı', 'AG Klemens Arızası', 'OG Sigorta Atması', 'AG Branşman Yeraltı Kablo Arızası',
       'AG Havai Branşman Arızası', 'AG Direkten Kesme Açma', 'AG Tel Kopuğu', 'AG Box / Sdk Abone Çıkış Sigorta Atığı',
       'AG Termik Açması', 'AG Box / Sdk Abone Çıkış Faz Sigorta Atığı', 'AG Sehim Bozukluğu', 'AG Pano Kol Sigorta Atığı', 
       'AG Pano Faz Sigorta Atığı', 'AG Atlama Kopuğu', 'AG Yük Ayırıcı Arızası', 'AG İzolatör Arızası', 'OG İzolatör Arızası',
       'OG Ayırıcı Arızası', 'OG İletken Kopması', 'AG Nötr İletken Kopması', 'AG Box / Sdk Giriş Faz Sigorta Atığı',
       'OG Atlama Arızası', 'OG Trafo Arızası', 'OG Kablo Başlığı Arızası', 'AG Voltaj Düşüklüğü', 'AG Kablo Başlığı Arızası',
       'OG Atlama(Jumper) Arızası', 'AG Box Arızası', 'OG Yeraltı Kablo Arızası', 'NH Altlık Arızası', 'AG Pano Arızası',
       'AG Box / Sdk Giriş Sigorta Atığı', 'OG Kesici Arızası', 'Sayaç Yanması', 'OG Parafudr Patlaması', 'OG Klemens Arızası',
       'AG Travers Arızası', 'TEİAŞ Trafo Arızası', 'AG Havai Branşman  Arızası', 'AG Voltaj Yüksekliği', 'AG Branşman Kablosu Arızası',
       'OG Travers Arızası', 'AG Yeraltı Kablo Arızası', 'AG Sehim Bozukluğu ve İletken Dolaşıklığı', 'EİH Arızası', 'AG Nötr İletkeni Kopması',
       'Frekans Düşmesi', 'AG Yeraltı Kablosu Arızası'
       
    ]
    #doğal sebepler
    natural_causes = [
        'Ağaç Kesimi', 'AG Hatta Ağaç Kesimi', 'AG Direk Kırılması', 'Yangın', 'Yabani Hayvan Teması', 'AG Hatta Yabancı Cisim',
        'OG Hatta Yabancı Cisim', 'OG Hatta Ağaç Kesimi', 'OG Direk Yıkılması' 
        
    ]
    #bakım çalışmaları
    maintenance = [
       'Şebeke Bakım Çalışması', 'Manevra', 'AG Direk Değişimi', 'Trafo Bakım Çalışması', 'OG Kademe Ayarı', 'AG Kademe Ayarı',
       'OG Trafo Kademe Ayarı', 'TEİAŞ Trafo Bakım Çalışması', 'OG Kademe ayarı ( OG / AG )', 'TEİAŞ Hat Bakım Çalışması', 
       'TEİAŞ İrtibat ve Yeni Hat Çalışması','OG Kademe ayarı ( OG / OG )','Yük Aktarımı'
        
       
    ]
    #3 şahış etkileri
    third_party_impacts = [
      'Müteahhit Çalışması', 'Üçüncü Şahısların Vermiş Olduğu Hasarlar', 'Fiziki İrtibat', 'Hırsızlık'
    ]
    if cause in technical_failures:
        return 'Teknik Arızalar'
    elif cause in natural_causes:
        return 'Doğal Nedenler'
    elif cause in maintenance:
        return 'Bakım Çalışmaları'
    elif cause in third_party_impacts:
        return 'Üçüncü Şahıs Etkileri'
    else:
        return 'Diğer'
```

```python
column_data = mugla_df['KESİNTİ NEDENİNE İLİŞKİN AÇIKLAMA (4)']
mugla_df['Kesinti_Nedeni_Kategori'] = column_data.apply(classify_cause)
print("Kategori frekansları:")
print(mugla_df['Kesinti_Nedeni_Kategori'].value_counts())
```
